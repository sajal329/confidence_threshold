```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Create a synthetic dataset
X, y = make_classification(n_samples=1000, n_features=3, n_informative=2, n_redundant=1, n_classes=2, n_clusters_per_class=1, random_state=42)

# Use only the first two features for visualization
X_visual = X[:, :2]

# Split data into labeled and unlabeled subsets
labeled_mask = np.random.choice([True, False], size=1000, p=[0.1, 0.9])
X_labeled = X[labeled_mask]
y_labeled = y[labeled_mask]
X_unlabeled = X[~labeled_mask]
y_unlabeled = y[~labeled_mask]  # For evaluation purposes only (not used in training)
```

```python
# Visualize the labeled and unlabeled data
plt.scatter(X_unlabeled[:, 0], X_unlabeled[:, 1], c='gray', label='Unlabeled Data', alpha=0.5)
plt.scatter(X_labeled[y_labeled == 0, 0], X_labeled[y_labeled == 0, 1], color='red', marker='x', label='Labeled Data (Class 0)', edgecolor='k')
plt.scatter(X_labeled[y_labeled == 1, 0], X_labeled[y_labeled == 1, 1], color='blue', marker='o', label='Labeled Data (Class 1)', edgecolor='k')
plt.title('Labeled vs. Unlabeled Data')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Train an initial model on labeled data
# Random Forest is chosen as the classifier for its robustness and ease of use
model = RandomForestClassifier(random_state=42)
model.fit(X_labeled, y_labeled)

# Generate pseudo-labels for unlabeled data with confidence scores
# Predict probabilities for the unlabeled dataset
probabilities = model.predict_proba(X_unlabeled)
pseudo_labels = model.predict(X_unlabeled)
confidence_scores = np.max(probabilities, axis=1)  # Extract the highest confidence score for each prediction
```

```python
# Plot the distribution of confidence scores for pseudo-labels
plt.figure(figsize=(8, 6))
plt.hist(confidence_scores, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(x=0.9, color='red', linestyle='--', label='Confidence Threshold (0.9)')
plt.title('Confidence Scores for Pseudo-Labels', fontsize=14)
plt.xlabel('Confidence Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Apply confidence thresholding
# Only retain predictions with confidence scores above the threshold
confidence_threshold = 0.9
high_confidence_mask = confidence_scores > confidence_threshold
X_high_confidence = X_unlabeled[high_confidence_mask]
y_high_confidence = pseudo_labels[high_confidence_mask]

# Combine labeled and high-confidence pseudo-labeled data
# Augment the labeled dataset with high-confidence pseudo-labeled data
X_combined = np.vstack((X_labeled, X_high_confidence))
y_combined = np.hstack((y_labeled, y_high_confidence))

# Retrain the model with the combined dataset
# Retrain the Random Forest model with the extended dataset
model_retrained = RandomForestClassifier(random_state=42)
model_retrained.fit(X_combined, y_combined)

# Evaluate the model
# Evaluate on the entire dataset for comparison
# Accuracy is computed for both the initial and retrained models
y_full_pred_initial = model.predict(X)
y_full_pred_retrained = model_retrained.predict(X)

initial_accuracy = accuracy_score(y, y_full_pred_initial)
retrained_accuracy = accuracy_score(y, y_full_pred_retrained)

print(f"Initial Model Accuracy: {initial_accuracy:.2f}")
print(f"Retrained Model Accuracy: {retrained_accuracy:.2f}")
```

```python
def plot_decision_boundary_with_annotations(model, X_visual, y, title):
    # Define grid boundaries for visualization
    x_min, x_max = X_visual[:, 0].min() - 1, X_visual[:, 0].max() + 1
    y_min, y_max = X_visual[:, 1].min() - 1, X_visual[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    # Add a dummy third feature to match model expectations
    grid_points = np.c_[xx.ravel(), yy.ravel(), np.zeros(xx.ravel().shape)]

    # Predict on the grid points
    Z = model.predict(grid_points)
    Z = Z.reshape(xx.shape)
    
    # Plot decision boundary
    plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')
    
    # Plot data points with markers and annotations
    plt.scatter(X_visual[y == 0, 0], X_visual[y == 0, 1], color='red', marker='x', label='Cluster 1 (Class 0)', edgecolor='k')
    plt.scatter(X_visual[y == 1, 0], X_visual[y == 1, 1], color='blue', marker='o', label='Cluster 2 (Class 1)', edgecolor='k')
    
    # Add axis labels, title, legend, and annotation
    plt.title(title, fontsize=14)
    plt.xlabel("Feature 1", fontsize=12)
    plt.ylabel("Feature 2", fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.legend(fontsize=10)
    plt.annotate("Cluster 1", xy=(-2, 0), xytext=(-3, 1), arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color='red')
    plt.annotate("Cluster 2", xy=(1, 1), xytext=(2, 2), arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color='blue')
    plt.show()

# Visualize the decision boundaries before and after retraining
plot_decision_boundary_with_annotations(model, X_visual, y, "Initial Model Decision Boundary")
plot_decision_boundary_with_annotations(model_retrained, X_visual, y, "Retrained Model Decision Boundary")
```
